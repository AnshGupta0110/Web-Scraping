```python
import requests
from bs4 import BeautifulSoup
```

```python
title = str(input('Enter the topic: ')).replace(' ','+')
link =  'https://www.google.com/search?q=' +title+'+wikipedia'


res = requests.get(link)
soup = BeautifulSoup(res.text,'html.parser')


for sp in soup.find_all('div'):
    try:
        link = sp.find('a').get('href')
        if('en.wikipedia.org' in link):
            break
    except:
        pass
link = link[7:].split('&')[0]
```

```python
res = requests.get(link)
soup = BeautifulSoup(res.text,'html.parser')
```

```python
heading = soup.find('h1').text
```

```python
heading
```

`data cleaning`

```python
corpus = ''

for p in soup.find_all('p'):
    corpus += p.text
    corpus += '\n'

corpus = corpus.strip()


print(corpus)
```

```python
for i in range(1,362):
    corpus = corpus.replace('['+str(i)+']','')
    
print(corpus)
```

`Creating text file name(heading), saving `

```python
fd = open(heading + '.txt', 'w', encoding = 'UTF-8')
fd.write(corpus)
fd.close()
```
